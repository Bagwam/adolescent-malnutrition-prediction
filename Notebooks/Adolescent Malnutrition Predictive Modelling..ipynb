```python
# Step 1: Clean the dataset by filtering only relevant disaggregation values (Boys and Girls)
cleaned_df = country_adolescent[
    country_adolescent['disagg.value'].isin(['Boys', 'Girls'])
]

# Step 2: Drop projection rows and unnecessary columns
cleaned_df = cleaned_df.drop(columns=['iso3', 'section'], errors='ignore')

# Step 3: Handle missing values by dropping columns with high null counts (e.g., >50% missing)
missing_threshold = 0.5
cleaned_df = cleaned_df.loc[:, cleaned_df.isnull().mean() < missing_threshold]

# Step 4: Group by gender to calculate global averages for each indicator (Obesity, Overweight, Thinness)
gender_trends = {
    "Boys": cleaned_df[cleaned_df['disagg.value'] == 'Boys']
        .filter(like="adolescent_")
        .mean(),
    "Girls": cleaned_df[cleaned_df['disagg.value'] == 'Girls']
        .filter(like="adolescent_")
        .mean()
}
```

# Feature Engineering 

```python
import pandas as pd

# Load the dataset to inspect its contents
file_path = 'C:/Users/User/Downloads/GNR_2021.xlsx.xlsx'
nutrition_data = pd.ExcelFile(file_path)

# Checking sheet names to understand the structure of the dataset
nutrition_data.sheet_names
```

```python
# Load the sheets relevant to Objective 2: preprocessing and feature engineering for predictive modeling
adolescent_country = nutrition_data.parse('Country adolescent')
adolescent_region = nutrition_data.parse('Region adolescent')
dietary_needs = nutrition_data.parse('Country dietary needs')
determinants = nutrition_data.parse('Country determinants')
policy_data = nutrition_data.parse('Country policy')

# Display basic info for understanding structure and relevant columns quickly
info_adolescent = adolescent_country.info()
info_dietary_needs = dietary_needs.info()
info_determinants = determinants.info()

# Show a brief preview of each sheet to identify key columns for feature engineering
preview_adolescent = adolescent_country.head()
preview_dietary_needs = dietary_needs.head()
preview_determinants = determinants.head()

info_adolescent, info_dietary_needs, info_determinants, preview_adolescent, preview_dietary_needs, preview_determinants
```

```python
# Step 1: Create engineered features for time-series trends (rolling averages and rate of change)
# Focus on adolescent obesity, overweight, and thinness trends

# Compute 3-year rolling averages for malnutrition indicators (smoothing fluctuations)
adolescent_country['obesity_3yr_avg'] = adolescent_country.loc[:, 'adolescent_obesity_2000':'adolescent_obesity_2019'].mean(axis=1)
adolescent_country['overweight_3yr_avg'] = adolescent_country.loc[:, 'adolescent_overweight_2000':'adolescent_overweight_2019'].mean(axis=1)
adolescent_country['thinness_3yr_avg'] = adolescent_country.loc[:, 'adolescent_thinness_2000':'adolescent_thinness_2019'].mean(axis=1)

# Compute percentage change in obesity/thinness rates between 2000 and 2019
adolescent_country['obesity_pct_change'] = ((adolescent_country['adolescent_obesity_2019'] - 
                                             adolescent_country['adolescent_obesity_2000']) / 
                                            adolescent_country['adolescent_obesity_2000']) * 100

adolescent_country['thinness_pct_change'] = ((adolescent_country['adolescent_thinness_2019'] - 
                                              adolescent_country['adolescent_thinness_2000']) / 
                                             adolescent_country['adolescent_thinness_2000']) * 100

# Step 2: Create dietary adequacy score (normalize nutrient intake and sum)
dietary_needs['dietary_diversity_score'] = dietary_needs[['Fruit', 'Vegetables', 'Legumes', 'Nuts', 'Whole grains', 'Fish', 'Dairy']].sum(axis=1)

# Normalize dietary diversity score by dividing by country-specific max for scaling purposes
dietary_needs['dietary_score_normalized'] = dietary_needs['dietary_diversity_score'] / dietary_needs['dietary_diversity_score'].max()

# Step 3: Create interaction terms from determinants data (e.g., education × sanitation)
determinants['education_sanitation_interaction'] = determinants['female_secondary_enroll_net_2019'] * determinants['surface_water_2019']

# Preview newly created features
engineered_features_preview = {
    'Adolescent Time-Series Features': adolescent_country[['country', 'obesity_3yr_avg', 'obesity_pct_change', 'thinness_3yr_avg']].head(),
    'Dietary Adequacy': dietary_needs[['country', 'dietary_score_normalized']].head(),
    'Determinants Interaction': determinants[['country', 'education_sanitation_interaction']].head()
}

engineered_features_preview
```

```python
# Step 1: Forward fill for time-series adolescent data (fill gaps in malnutrition indicators)
adolescent_country_ffill = adolescent_country.copy().fillna(method='ffill', axis=1).fillna(method='bfill', axis=1)

# Step 2: Regional median imputation for dietary needs data
dietary_needs_imputed = dietary_needs.copy()
for column in ['Fruit', 'Vegetables', 'Legumes', 'Nuts', 'Whole grains', 'Fish', 'Dairy']:
    dietary_needs_imputed[column].fillna(dietary_needs.groupby('region')[column].transform('median'), inplace=True)

# Step 3: Initialize determinants_imputed and apply iterative imputation
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Create a copy of determinants data
determinants_imputed = determinants.copy()

# Impute only critical columns to speed up the process and avoid long computations
critical_columns = ['female_secondary_enroll_net_2019', 'surface_water_2019']

# Apply IterativeImputer to the selected subset
imputer = IterativeImputer(max_iter=10, random_state=0)
determinants_imputed[critical_columns] = imputer.fit_transform(determinants_imputed[critical_columns])

# Verify the imputed results across datasets
adolescent_ffill_preview = adolescent_country_ffill[['country', 'obesity_3yr_avg', 'obesity_pct_change']].head()
dietary_imputed_preview = dietary_needs_imputed[['country', 'dietary_score_normalized']].head()
determinants_imputed_preview = determinants_imputed[['country', 'female_secondary_enroll_net_2019']].head()

adolescent_ffill_preview, dietary_imputed_preview, determinants_imputed_preview
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import mutual_info_regression
import numpy as np

# Step 1: Prepare the merged dataset (adolescent, dietary, and determinants data)
# Merge key features into a single dataset for selection
# Merge datasets
merged_data = pd.merge(adolescent_country_ffill[['country', 'obesity_3yr_avg', 'obesity_pct_change', 'thinness_3yr_avg']],
                       dietary_needs_imputed[['country', 'dietary_score_normalized']], on='country', how='inner')
merged_data = pd.merge(merged_data, determinants_imputed[['country', 'female_secondary_enroll_net_2019']], on='country', how='inner')
```

```python
# Drop country column and keep only numeric data
X_final = merged_data.drop(columns=['country']).apply(pd.to_numeric, errors='coerce').dropna()
```

```python
# Set target variable
target = 'obesity_3yr_avg'

# Initialize Random Forest Regressor
rf = RandomForestRegressor(random_state=0)

# Fit the model
rf.fit(X_final.drop(columns=[target]), X_final[target])

# Rank features based on importance
feature_importances = pd.Series(rf.feature_importances_, index=X_final.drop(columns=[target]).columns).sort_values(ascending=False)

# Display top 5 features
feature_importances.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier
import pandas as pd
import numpy as np

# Simulated dataset (expand with actual data locally if needed)
np.random.seed(42)
simulated_data = {
    'thinness_3yr_avg': np.random.uniform(8, 20, 500),
    'obesity_pct_change': np.random.uniform(10, 35, 500),
    'dietary_score_normalized': np.random.uniform(0.4, 0.8, 500),
    'female_secondary_enroll_net_2019': np.random.uniform(50, 95, 500),
    'policy_effectiveness_score': np.random.uniform(1, 5, 500),
    'gdp_per_capita': np.random.uniform(1000, 50000, 500),
    'poverty_rate': np.random.uniform(5, 50, 500),
}

# Create DataFrame and target variable
expanded_df = pd.DataFrame(simulated_data)
expanded_df['obesity_risk'] = (expanded_df['obesity_pct_change'] > np.median(expanded_df['obesity_pct_change'])).astype(int)

# Split into features and target
X_expanded = expanded_df.drop(columns=['obesity_risk'])
y_expanded = expanded_df['obesity_risk']

# Train-test split
X_train_exp, X_test_exp, y_train_exp, y_test_exp = train_test_split(X_expanded, y_expanded, test_size=0.2, random_state=42)

# Initialize XGBoost with the best parameters
best_xgb = XGBClassifier(
    learning_rate=0.01, 
    max_depth=3, 
    n_estimators=50, 
    subsample=0.6, 
    eval_metric='logloss', 
    random_state=0
)

# Train and predict
best_xgb.fit(X_train_exp, y_train_exp)
y_pred_exp = best_xgb.predict(X_test_exp)

# Evaluate accuracy and classification performance
final_accuracy = accuracy_score(y_test_exp, y_pred_exp)
final_report = classification_report(y_test_exp, y_pred_exp, target_names=['Low Obesity Risk', 'High Obesity Risk'])

print("Final Test Set Accuracy:", final_accuracy)
print("Classification Report:\n", final_report)
```

```python
from scipy.stats import ttest_ind

# Simulate predictor values for high-income and low-/middle-income regions
np.random.seed(42)

high_income_data = {
    'thinness_3yr_avg': np.random.uniform(8, 15, 250),
    'obesity_pct_change': np.random.uniform(20, 35, 250),
    'dietary_score_normalized': np.random.uniform(0.5, 0.8, 250),
    'female_secondary_enroll_net_2019': np.random.uniform(80, 95, 250)
}

low_mid_income_data = {
    'thinness_3yr_avg': np.random.uniform(12, 20, 250),
    'obesity_pct_change': np.random.uniform(10, 25, 250),
    'dietary_score_normalized': np.random.uniform(0.4, 0.7, 250),
    'female_secondary_enroll_net_2019': np.random.uniform(50, 80, 250)
}

# Create DataFrames
high_income_df = pd.DataFrame(high_income_data)
low_mid_income_df = pd.DataFrame(low_mid_income_data)

# Perform t-tests between regions
t_test_results = {
    feature: ttest_ind(high_income_df[feature], low_mid_income_df[feature], equal_var=False)
    for feature in high_income_df.columns
}

# Display t-statistics and p-values
t_test_summary = {feature: {"t-statistic": result.statistic, "p-value": result.pvalue} for feature, result in t_test_results.items()}
print(pd.DataFrame(t_test_summary))
```
